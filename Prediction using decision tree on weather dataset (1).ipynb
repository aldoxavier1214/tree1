##### Dropping The Unnecessary column as it contains high null values 

```python
df=df.drop(['Date', 'Location', 'Evaporation','Sunshine', 'Cloud9am', 'Cloud3pm'],axis=1)
df.head()
```

##### Dropping null values in the rows

```python
df=df.dropna(axis=0)
df.shape
```

```python
df.columns
```

##### Categorical column

```python
catcols=df.select_dtypes(include='object').columns
for i in catcols:
    print(i)
```

##### Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in catcols:
    df[i]=le.fit_transform(df[i])
```

```python
df.dtypes
```

```python
X=df.drop(columns=["RainTomorrow"],axis=1)
y=df["RainTomorrow"]
```

##### Scatterplot

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='MaxTemp',y='MinTemp',hue='RainTomorrow',palette='inferno',data=df);
```

When min temp increases max temp is also increasing and this is in linear form but it is a cluster

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='Humidity9am',y='Temp9am',hue='RainTomorrow',palette='inferno',data=df);
```

As the humidity is increasing the temperature and probability of getting rain is also increasing

##### HeatMap

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),annot=True);
```

RainTomorrow has good relationship with humidity9am , humidity3pm and raintoday as light colour shows the good relationship in the heatmap.

##### Splitting the dataset for model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.metrics import classification_report ,confusion_matrix ,accuracy_score
```

##### Model building Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="gini",min_samples_split=2)
dt.fit(X_train,y_train)
score=dt.score(X_test,y_test)
print(f'Model accuracy on test ={score*100:.2f}%')
```

##### Model Evaluation

```python
y_pred=dt.predict(X_test)
print(f"Classification report:\n\n {classification_report(y_test,y_pred)}")
```

```python
dt_matrix = confusion_matrix(y_test,y_pred)
print(f"Confusion Matrix:\n\n{dt_matrix}")
```

##### Visualization of Confusion matrix

```python
plt.subplots
sns.heatmap(dt_matrix,annot=True,cmap='coolwarm',fmt='.2f');
plt.xlabel='Predicted label'
plt.ylabel='Actual label'
```

```python
print(f"Accuracy Score:\n\n{accuracy_score(y_test,y_pred)*100:.2f}%")
```
